```python
from matplotlib import pyplot
import random
```

```python
def unfairness(job_length):
    time = 0
    win_counts = [0, 0]
    while True:
        time += 1
        this_winner = random.randint(0, 1)
        win_counts[this_winner] += 1
        if win_counts[this_winner] == job_length:
            winner = this_winner
            loser = 1 - this_winner
            loser_remaining_time = job_length - win_counts[loser]
            return time/(time+loser_remaining_time)


unfairnesses = []
experiment_count = 10

for job_length in range(1, 1000):
    unfairness_avg = 0
    for _ in range(experiment_count):
        unfairness_avg += unfairness(job_length)
    unfairness_avg = unfairness_avg / experiment_count
    unfairnesses.append(unfairness_avg)
```

```python
pyplot.plot(unfairnesses)
pyplot.xlabel("job length")
pyplot.ylabel("Unfairness (average of {} times)".format(experiment_count))
```
